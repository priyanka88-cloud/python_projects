```python
seaborn.relplot(data=data,kind='line',x='body_mass_g',y='flipper_length_mm',hue='species') #as bodymass increses flipper length increases
```

```python
#distribution of data called Histogram
seaborn.displot(data,x='body_mass_g')#seaborn.displot, a function for creating distribution plots in the Seaborn library.
#when you only specify the x parameter, the default y-axis represents the frequency (or count) of the values in the x-axis variable.
#x-numeric data column
#y-count
#vertical bar can change
#penguin having flipper_length_mm is between 190 to 195 is appear more than other
#more penguin has  flipper_length_mm is between 190 to 195
```

```python
seaborn.displot(data,x='body_mass_g',hue='sex',multiple='dodge')#The bins parameter in seaborn.displot specifies the number of bins to use in the histogram or the bin edges.
# By default, Seaborn automatically determines the number of bins,
#but you can customize it .
```

```python
seaborn.displot(data,x='flipper_length_mm',binwidth=10) #binwidth parameter in seaborn.displot allows you to specify the width of each bin,
```
