```python
'''
The groupby method in pandas is a powerful tool for splitting a DataFrame into groups,
 performing operations on those groups, and then combining the results.
 here groupby method split 'data' into group 'sex' and perform all the aggregate operations on the 'sex' group
 here it is important to make group bez aggregate operations are perform on groups only
#aggrgate function
#data.groupby('categorical column name').agg(
#{ 'numerical column name':['aggreage function name']}(options are'mean',''max','min','std','count','sum')
#)
'''
data.groupby('sex').agg({
    'total_bill':['mean','max','min','std','count','sum'],
    'tip':['mean','max'],
    'size':'mean'
})
```

```python
'''
A pivot table is a data summarization tool commonly used in data processing and analysis.represent in table format
Pivot tables are highly useful for transforming large datasets into a more understandable and useful
 summary by breaking down the data into various dimensions.
#pandas.pivot_table(dataframe name,index='caterorical col',column='cal cloumn',values='numeric col',aggfun='agg function)

'''
pandas.pivot_table(data,index='day',columns='sex',values='total_bill',aggfunc='mean')
'''
#pandas.pivot_table function is used to create a spreadsheet-style pivot table as a DataFrame.
data: DataFrame you want to pivot.
index: Column(s) to set as index. means which rows u want
columns: Column(s) to set as columns.means which column u want
values: Column(s) to aggregate. on which column u want to peeform aggreagae function here is 'mean'
so rows are day and columns are sex that is male and female and values or data of them(sex and days ) are total_bill and aggfun perform mean on total_bill
aggfunc: Function(s) to use for aggregation (default is 'mean').

'''
```

```python
'''
Functional Magnetic Resonance Imaging (fMRI) is a neuroimaging technique used to
measure and map brain activity.
Unlike traditional MRI, which provides static images of brain structure,
 fMRI captures dynamic brain activity by detecting changes in blood flow.
 #timepoint represents the time points at which brain activity was recorded.
 event= stim or cue
 region= The brain region where the signal was measured.
 timepoint=represents the time points at which brain activity was recorded.
 signal=represents the recorded brain activity.

'''
fmri=seaborn.load_dataset('fmri')
fmri.head()
```

```python
seaborn.relplot(kind='line',x='timepoint',y='signal',data=fmri) # here the brain signal vary with certain time(timepoint)
```

```python
seaborn.relplot(kind='line',x='timepoint',y='signal',data=fmri,style='event')
# researchers often design experiments where participants are exposed to various stimuli (such as images, sounds, or tasks) to see how different brain regions respond.
#a cue (such as a signal or instruction) is given to the participant, and their brain activity is recorded in response to it.
```
